```python
def download_lyrics(artist, title):
    
    while True:
        try:
            song = genius.search_song(title, artist)
            #song_lyrics = song.lyrics.split()
            if song is not None:
                return song.lyrics
        except:
            pass
        
```

```python
for ind in df.index:
    artist = df['Artist'][ind]
    title = df['Title'][ind]
    lyrics = download_lyrics(artist, title)
    df.loc[ind, "Lyrics"] = lyrics
```
