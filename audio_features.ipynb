```python
import set_env
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
from pprint import pprint
from pathlib import Path
```

```python
spotify = spotipy.Spotify(client_credentials_manager=SpotifyClientCredentials())
```

```python
csv_path = Path("top_ten_songs.csv")
df = pd.read_csv(csv_path)
df
```

```python
#song_names = df['Title'].to_list()
```

```python
track_result = []

for ind in df.index:
    song = df['Title'][ind]
    track_data = spotify.search(q='track:'+ song, type='track', limit = 1)
    max_pop = 0
    max_pop_id = 0
    for i, item in enumerate(track_data['tracks']['items']):
        popularity = item['popularity']
        if popularity > max_pop:
            popularity = max_pop
            max_pop_id = i  
    track_name = df['Title'][ind]
    popularity = track_data['tracks']['items'][max_pop_id]['popularity']
    track_id = track_data['tracks']['items'][max_pop_id]['id']
    artist = df['Artist'][ind]
    track_result.append((artist, track_id, track_name, popularity))
#pprint(track_data)
```

```python
track_result
```

```python
track_df = pd.DataFrame(track_result, index = None, columns=('Artist','id',
                                                             'Song Name','Popularity'))

track_df
```

```python
features_df =  pd.DataFrame(columns=['id','Danceability','Energy','Tempo','Duration'])

for album in range(0,180):
    features = spotify.audio_features(track_df['id'][album])
    danceability = features[0]['danceability']
    energy = features[0]['energy']
    tempo = features[0]['tempo']
    duration = features[0]['duration_ms']
    track_id = features[0]['id']
    
    features_df.loc[len(features_df.index)] = [track_id, danceability, energy, tempo, duration]  

features_df
```

```python
track_df['Danceability'] = features_df['Danceability']
track_df['Energy'] = features_df['Energy']
track_df['Tempo'] = features_df['Tempo']
track_df['Duration'] = features_df['Duration']
track_df
```

```python
track_df.to_csv("output.csv")
```
