```python
top_songs_df =  pd.DataFrame(columns=['Year','Title','Artist'])
for year_iter in range(2006, 2024):
    data = billboard.ChartData("hot-100-songs", year=year_iter)
    for song in range(1):
        title = data.entries[song].title
        artist = data.entries[song].artist
        top_songs_df.loc[len(top_songs_df.index)] = [year_iter, title, artist]  

top_songs_df
```

```python
top_ten_songs_df =  pd.DataFrame(columns=['Year','Rank','Title','Artist'])
for year_iter in range(2006, 2024):
    data = billboard.ChartData("hot-100-songs", year=year_iter)
    for song in range(0,10):
        rank = data.entries[song].rank
        title = data.entries[song].title
        artist = data.entries[song].artist
        top_ten_songs_df.loc[len(top_ten_songs_df.index)] = [year_iter, rank, title, artist]  

top_ten_songs_df
```

```python
from pathlib import Path  
filepath = Path('top_songs.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
top_songs_df.to_csv(filepath)  
```

```python
top_hundred_songs_df =  pd.DataFrame(columns=['Year','Rank','Title','Artist'])
for year_iter in range(2006, 2024):
    data = billboard.ChartData("hot-100-songs", year=year_iter)
    for song in range(0,99):
        rank = data.entries[song].rank
        title = data.entries[song].title
        artist = data.entries[song].artist
        top_hundred_songs_df.loc[len(top_hundred_songs_df.index)] = [year_iter, rank, title, artist]  

top_hundred_songs_df
```

```python
categories_df =  pd.DataFrame(columns=['Category','Rank','Title','Artist'])
categories = ['greatest-of-all-time', 'pop','country', 'rock', 'alternative','r-b-hip-hop', 'dance-electronic', 'latin']
for cat in categories: 
    data = billboard.ChartData(f'{cat}-song')
    for song in range(0,40):
        rank = data.entries[song].rank
        title = data.entries[song].title
        artist = data.entries[song].artist
        categories_df.loc[len(categories_df.index)] = [cat, rank, title, artist]  

categories_df 
```

```python
from pathlib import Path  
filepath = Path('categories_top_40.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
categories_df.to_csv(filepath)  
```
